```python
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path

BASE_DIR = Path.cwd()
data_file = BASE_DIR / "Timeseries_3.156_101.691_E5_1kWp_crystSi_14_0deg_-1deg_2023_2023.csv"

```

```python
arr = np.loadtxt(data_file, 
                 delimiter=',', 
                 skiprows=23, 
                 usecols=(1), 
                 max_rows=8760)

# arr = np.loadtxt('C:\\Users\\royro\\Desktop\\Solar_PPAs\\DraftReport15_01_26\\Timeseries_3.156_101.691_E5_1kWp_crystSi_14_0deg_-1deg_2023_2023.csv', 
#                  delimiter=',', 
#                  skiprows=23, 
#                  usecols=(1), 
#                  max_rows=8760)

solar_profile = (1/1000)*np.concatenate((np.zeros(8), arr[:-8]))

print(solar_profile.mean())
print(solar_profile.max())

avg_profile = solar_profile.reshape((365, 24)).mean(axis=0)

daily_profile = solar_profile.reshape((365, 24)).sum(axis=1)


plt.title("Average Daily Solar Profile")
plt.xlabel("Hour of Day")
plt.ylabel("PV production [kW/kWp]")
plt.grid()
plt.plot(avg_profile)

plt.show()

plt.title("Distribution of Daily Average Solar Irradiance")
plt.xlabel("Daily PV production [kWh]")
plt.ylabel("Frequency")
plt.hist(daily_profile, bins=15)
plt.grid()
plt.show()

plt.violinplot(daily_profile)
plt.title("Distribution of Daily Average Solar Irradiance")
plt.ylabel("Daily PV production [kWh]")
plt.grid()
plt.show()


plt.plot(solar_profile, '.')
plt.show()
```
